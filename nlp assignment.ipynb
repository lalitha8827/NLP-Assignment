# In this assignment students have to find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

import necessary libraries

```python
from bs4 import BeautifulSoup
import urllib.request as url
import nltk
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

extracting text from url

```python
response=url.urlopen('http://php.net/')
html=response.read()
soup=BeautifulSoup(html,"html5lib")
```

```python
text=soup.get_text()
```

extracting words from text

```python
words=re.findall('[a-zA-Z]+',text)
```

converting all uppercase letters to lowercase because python is case-sensitive

```python
list=[i.lower() for i in words]
```

```python
list
```

removal of stop words

```python
#stopwords are the words which occur frequently like a,the,an
```

```python
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
```

words after removal of stopwords

```python
final_words=[i for i in list if i not in stopwords]
```

frequency of all the words

```python
freq=nltk.FreqDist(final_words)
```

```python
for i,j in freq.items():
    print(i,":",j)
```

sorted values according to frequency

```python
sorted_values=sorted(freq.items(),key=lambda x:x[1],reverse=True)
for i,j in sorted_values:
    print(i,":",j)
```

```python
freq.plot(20)
plt.show()
```
